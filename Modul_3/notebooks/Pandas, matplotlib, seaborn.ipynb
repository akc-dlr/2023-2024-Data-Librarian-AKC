```python
# import the library and give it a shorter name 'pd'
import pandas as pd

# create a dataframe by hand with two columns and three rows
df = pd.DataFrame({
        'month': [1, 2, 3, 4, 5, 6, 7],
        'temperatur': [-12, 3, 9, 10, -10, 28, 4]
})

# print out some descriptive statistics
df.describe()
```

Welche Statistiken liefert ein Aufruf der Funktion describe()?

- Anzahl der Einträge (count): Anzahl der nicht fehlenden Werte in jeder Spalte.
- Mittelwert (mean): Durchschnittlicher Wert der Daten in jeder Spalte.
- Standardabweichung (std): Maß dafür, wie weit die Werte um den Mittelwert herum streuen.
- Minimum (min): Kleinster Wert in jeder Spalte.
- Quantile (25%, 50%, 75%): Werte, die die angegebene Prozentzahl der Daten überschreiten. Das 50%-Quantil ist auch der Median.
- Maximum (max): Größter Wert in jeder Spalte.
