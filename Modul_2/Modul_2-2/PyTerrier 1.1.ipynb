```python
import pyterrier as pt
import requests
import pandas as pd
import pickle
import nltk
import numpy as np
```

```python
# json in dataframe umwandeln
```

```python
import json
```

```python
data = json.loads(open("/home/anni/Schreibtisch/2023-2024-Data_Librarian_Ann-Kathrin_Christann/Modul_2/Modul_2-1/jsonoutput.json").read())
```

```python
df = pd.DataFrame(data)
```

```python
# pyterrier initialisieren

if not pt.started():
    pt.init()

```

```python
# docno und text Feld 
```

```python
df
```

```python
df['docno'] = df['source']
df['text'] = df['title']
```

```python
df
```

```python
# Daten indexieren mit mehreren Feldern

# dataframe in Liste auf dicts umwandel

df_dict = df.to_dict(orient='records')
```

```python
index_folder = "/home/anni/Schreibtisch/2023-2024-Data_Librarian_Ann-Kathrin_Christann/Modul_2/Modul_2-1/df_index"

fields=['docno', 'text', 'authors', 'searchterm', 'publication_year']

indexer_folder = pt.IterDictIndexer(index_folder, meta={'docno': 200, 'text': 4096}, overwrite=True)
index_ref = indexer_folder.index(df_dict, fields=fields)

print(f"path to our index: {index_ref.toString()}")
```

```python
index = pt.IndexFactory.of(index_ref)

```

```python
print(index.getCollectionStatistics().toString())
```

```python
# Suchmaschine bauen

search_engine = pt.BatchRetrieve(index, wmodel="TF_IDF")
```

```python
res = search_engine.search("star")
res
```
